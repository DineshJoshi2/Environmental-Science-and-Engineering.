# (1) The mass of TSS sludge in the system (MXTSS, kgTSS):
 
 To calculate the masses of volatile 
(MXv) and total (MXt) settleable solids in the system
for the raw wastewater and settled waste water we need Endogenous respiration rate $(\frac 1d)$ $b_H$ which is temperature dependence. Temperature dependency of the Endogenous respiration rate ($b_H$) is show by the equation $b_{HT} = b_{H20}  θ^{(T-20)}$. Where the value for $θ$ is 1.029 and the standard value at $20^o$ for $b_{H20}$ is 0.24.

Sludge ages ranges from 3 to 30 
days.

Here,

MXv_Raw_14 : Mass of VSS at temperature $14^o$ C for raw waste water

MXt_Raw_14 : Mass of TSS at temperature $14^o$ C for raw waste water

MXv_Raw_22 : Mass of VSS at temperature $22^o$ C for raw waste water

MXt_Raw_22 : Mass of TSS at temperature $22^o$ C for raw waste water

MXv_Settled_14 : Mass of VSS at temperature $14^o$ C for settled waste water

MXt_Settled_14 : Mass of TSS at temperature $14^o$ C for settled waste water

MXv_Settled_22 : Mass of VSS at temperature $22^o$ C for settled waste water

MXt_Settled_22 : Mass of TSS at temperature $22^o$ C for settled waste water
